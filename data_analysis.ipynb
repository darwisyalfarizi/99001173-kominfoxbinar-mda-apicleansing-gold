## II. Data Preprocessing


### Data cleaning

````python
print(df['Tweet'].dtype)
````

````python
df['Tweet'] = df['Tweet'].astype(str)

````

````python
print(df['Tweet'].dtype)
````

#### Replace  kamus alay word to standard word

````python
dictionary = dict(zip(df_kamusalay['anakjakartaasikasik'], df_kamusalay['anak jakarta asyik asyik']))
df['Tweet'] = df['Tweet'].apply(lambda x: " ".join([dictionary.get(w, w) for w in x.split()]))
````

````python
df
````

#### Hapus stopword


````python
df
````

````python
#remove stopword
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

nltk.download('stopwords')
nltk.download('punkt')
nltk.corpus.stopwords.words('indonesian')

# tokenize
df['tweet_tokenized'] = df['Tweet'].apply(lambda x: word_tokenize(x))

# Removing common words that usually have no important meaning (stop words)
stop_words = stopwords.words("indonesian")

# create 'tweet_without_stopwords' 
df['tweet_without_stopwords'] = df['tweet_tokenized'].apply(lambda x: ' '
.join([word for word in x if word not in stop_words]))
````

````python
df.head()
````

````python
#indexing
df = df[['Tweet',
 'tweet_tokenized',
 'tweet_without_stopwords',
 'HS',
 'Abusive',
 'HS_Individual',
 'HS_Group',
 'HS_Religion',
 'HS_Race',
 'HS_Physical',
 'HS_Gender',
 'HS_Other',
 'HS_Weak',
 'HS_Moderate',
 'HS_Strong',
 ]]
````

````python
df.head()
````

````python
# add label 

df['Abusive_Label'] = df['Abusive'].apply(lambda x: 'Abusive' if x == 1 else 'Non Abusive')
````

````python
df
````

````python
df['total_char_wo_stopwords'] = df['tweet_without_stopwords'].apply(len)
````

````python
# Melakukan Indexing Ulang dari df['tweet_without_stopwords]
df = df[['Tweet',
 'tweet_tokenized',
 'tweet_without_stopwords',
 'total_char_wo_stopwords', 'Abusive_Label',
 'HS',
 'Abusive',
 'HS_Individual',
 'HS_Group',
 'HS_Religion',
 'HS_Race',
 'HS_Physical',
 'HS_Gender',
 'HS_Other',
 'HS_Weak',
 'HS_Moderate',
 'HS_Strong',
 ]]
````

````python
df.head()
````

````python
df['total_word_wo_stopwords'] = df['tweet_without_stopwords'].apply(lambda sent: len(sent.split()))
````

````python
# Melakukan Indexing Ulang dari df['tweet_without_stopwords]
df = df[['Tweet',
 'tweet_tokenized',
 'tweet_without_stopwords',
 'total_char_wo_stopwords',
 'total_word_wo_stopwords', 'Abusive_Label',
 'HS',
 'Abusive',
 'HS_Individual',
 'HS_Group',
 'HS_Religion',
 'HS_Race',
 'HS_Physical',
 'HS_Gender',
 'HS_Other',
 'HS_Weak',
 'HS_Moderate',
 'HS_Strong',
 ]]
````
